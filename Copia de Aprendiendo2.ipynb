```python
a = ["Lucía"] * 70

for i in range(30):
    a.append("Lilia")
    a.append("Arturo")

familiares = ["padre", "madre", "comprensión", "hija", "abuelo", "abuela", "nieta", "madrina", "padrino", "hogar", "casa", "linaje", "descendencia", "ancestros", "parentesco", "consanguinidad", "afinidad", "casa", "amor", "cariño", "apoyo", "unión", "respeto", "lealtad", "confianza", "felicidad", "seguridad", "convivir", "compartir", "celebrar", "ayudarse", "jugar", "viajar", "conversar", "aprender", "crecer", "cuidarse", "hogar"]

for _ in range(np.random.randint(0, 500)):
    random_index = np.random.randint(0, len(familiares))
    a.append(familiares[random_index])

text = " ".join(a)

wc = WordCloud(background_color="white", max_words=1000, colormap="winter")

wc.generate(text)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 equally spaced points between 0 and 2*pi
x = np.linspace(0, 2*np.pi, 100)

# Calculate the sine of each point
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
a = [np.random.randint(-100, 100) for i in enumerate(range(15))]
b = [np.random.randint(-100, 100) for i in enumerate(range(15))]
c = [np.random.randint(-100, 100) for i in enumerate(range(15))]
letras = ["MXN", "USD", "UDI"]
df = pd.DataFrame(data=[a, b, c], index=letras)
df.loc['OTH'] = np.random.randint(1, 100, size=15)
df.loc['Total'] = df.sum()
df.loc['Bucket'] = [str(i) + 'Y' for i in range(1, 16)]
df.columns = df.iloc[5]
df = df.iloc[1:]
df.drop(df.loc['Bucket'].name, inplace=True)

# Gráfico
fig, ax = plt.subplots()

# Barras
barWidth = 0.5
bars1 = df.iloc[0].tolist()
bars2 = df.iloc[1].tolist()
bars3 = df.iloc[2].tolist()
bars4 = df.iloc[3].tolist()

# Posiciones de las barras en el eje x
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Graficar barras
plt.bar(r1, bars1, color='b', width=barWidth, edgecolor='grey', label='MXN')
plt.bar(r2, bars2, color='r', width=barWidth, edgecolor='grey', label='USD')
plt.bar(r3, bars3, color='g', width=barWidth, edgecolor='grey', label='UDI')
plt.bar(r4, bars4, color='y', width=barWidth, edgecolor='grey', label='OTH')

# Etiquetas de eje x
plt.xlabel('Bucket', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], df.columns)

# Etiquetas de eje y
plt.ylabel('Valor', fontweight='bold')

# Título del gráfico
plt.title('Gráfico de barras')

# Mostrar leyenda
plt.legend()

# Mostrar gráfico
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Datos
a = [np.random.randint(1, 100) for _ in range(15)]
b = [np.random.randint(1, 100) for _ in range(15)]
c = [np.random.randint(1, 100) for _ in range(15)]
letras = ["MXN", "USD", "UDI"]
df = pd.DataFrame(data=[a, b, c], index=letras)
df.loc['OTH'] = np.random.randint(1, 100, size=15)
df.loc['Total'] = df.sum()
df.loc['Bucket'] = [str(i) + 'Y' for i in range(1, 16)]
df.columns = df.iloc[5]
df = df.iloc[1:]
df.drop(df.loc['Bucket'].name, inplace=True)

# Transponer para tener las columnas como categorías
df = df.T.reset_index()

# Gráfico con Seaborn
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Utilizamos melt para convertir las columnas de letras en una variable categórica
sns.barplot(data=df.melt(id_vars='Bucket', var_name='Letra', value_name='Valor'),
            x='Bucket', y='Valor', hue='Letra', palette="bright")

plt.xlabel('Bucket', fontweight='bold')
plt.ylabel('Valor', fontweight='bold')
plt.title('Gráfico de barras con Seaborn')

plt.legend(title='Letra')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
